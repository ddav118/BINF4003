```python
import pandas as pd
import os
```

```python
def add_cui_compare(n, manual_annotation_dir, metamap_dir):

# List all files in each directory
    manual_annotation_files = os.listdir(manual_annotation_dir)
    metamap_files = os.listdir(metamap_dir)

# Loop through all the files in both directories
    for filename in manual_annotation_files:
        if filename in metamap_files:
            # Read CSV files into pandas dataframes
            manual_df = pd.read_csv(os.path.join(manual_annotation_dir, filename))
            metamap_df = pd.read_csv(os.path.join(metamap_dir, filename))

            # Add a new column called "confusion label" to metamap dataframe
            metamap_df['confusion label'] = -1
            manual_df['confusion label'] = -1   

            # Loop through all rows in metamap dataframe where negex = 0 (change to = 1 for neg analysis)
            for index, row in metamap_df[metamap_df['negex'] == n].iterrows():
                cui = row['cui'] 
                if any(manual_df['negex'] == n): # (change to = 1 for neg analysis)
                    # Check if cui exists in manual annotation dataframe
                    if cui in manual_df['cui'].values:
                        metamap_df.at[index, 'confusion label'] = 'TP'
                    else:
                        metamap_df.at[index, 'confusion label'] = 'FP'

            # Loop through all rows in manual annotation dataframe
            for index, row in manual_df[manual_df['negex'] == n].iterrows():
                cui = row['cui'] 
                if any(manual_df['negex'] == n): # (change to = 1 for neg analysis)
                # Check if cui does not exist in metamap annotation dataframe
                    if cui not in metamap_df['cui'].values: 
                        manual_df.at[index, 'confusion label'] = 'FN'

            # Save the updated dataframes with the new column
            metamap_df.to_csv(os.path.join(metamap_dir, filename), index=False)
            manual_df.to_csv(os.path.join(manual_annotation_dir, filename), index=False)
```

```python
# calculate precision, recall, F1-score
# count the total number of TP(meta_df)/ FP(meta_df)/ FN(manual_df) 

def cal(meta_directory, manual_directory):

    total_tp = 0
    total_fp = 0
    total_fn = 0

    # Loop through all files in the meta directory
    for filename in os.listdir(meta_directory):
        meta_df = pd.read_csv(os.path.join(meta_directory, filename))
        
        tp_count = len(meta_df[meta_df['confusion label'] == 'TP'])
        fp_count = len(meta_df[meta_df['confusion label'] == 'FP'])
        
        total_tp += tp_count
        total_fp += fp_count

    # Loop through all files in the manual directory
    for filename in os.listdir(manual_directory):
        manual_df = pd.read_csv(os.path.join(manual_directory, filename))
        if 'confusion label' in manual_df.columns:
            fn_count = len(manual_df[manual_df['confusion label'] == 'FN'])
            total_fn += fn_count

    # Print or use the total counts as needed
    print(f"Total TP: {total_tp}")
    print(f"Total FP: {total_fp}")
    print(f"Total FN: {total_fn}")

    precision = total_tp / (total_tp + total_fp)
    recall = total_tp / (total_tp + total_fn)
    f = 2 * (precision * recall) / (precision + recall)

    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f}")
```
