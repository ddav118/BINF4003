```python
#cohen kappa for interannotator before annotation guideline

tp = 0
tn = 0
fp = 0
fn = 0

filepath = '/Users/elizabethchang/Desktop/symbolic-project-data/interannotator'
file_list = os.listdir(os.path.join(filepath, 'cindy'))

for file in file_list:
    cc = pd.read_csv(os.path.join(filepath, 'cindy', file))
    cc_cui = cc['cui'].dropna().to_list()

    ec = pd.read_csv(os.path.join(filepath, 'elizabeth', file))
    ec_cui = ec['cui'].dropna().to_list()

    metamap = pd.read_csv(os.path.join(filepath,'metamap-csv', file))
    metamap_cui = metamap['cui'].dropna().to_list()

    cc_counter = Counter(cc_cui)
    ec_counter = Counter(ec_cui)
    metamap_counter = Counter(metamap_cui)

    for cui in metamap_counter:
        if cui not in cc_counter and cui not in ec_counter:
            tn += metamap_counter[cui]
    for cui in cc_counter:
        if cui in ec_counter:
            if cc_counter[cui] == ec_counter[cui]:
                tp += cc_counter[cui]
            elif cc_counter[cui] < ec_counter[cui]:
                tp += cc_counter[cui]
                fp += (ec_counter[cui] - cc_counter[cui])
            elif cc_counter[cui] > ec_counter[cui]:
                tp += ec_counter[cui]
                fn += (cc_counter[cui] - ec_counter[cui])
        else:
            fn += cc_counter[cui]
    for cui in ec_counter:
        if cui not in cc_counter:
            fp += ec_counter[cui]
            

print(f"Total TP: {tp}")
print(f"Total TN: {tn}")
print(f"Total FP: {fp}")
print(f"Total FN: {fn}")

precision = tp / (tp + fp)
recall = tp / (tp + fn)
f = 2 * (precision * recall) / (precision + recall)
cohen = (2 * (tp * tn - fn * fp))/((tp+fp)*(fp+tn)+(tp+fn)*(fn+tn))
per_match = (tp + tn)/(tp + tn + fp + fn)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f}")
print(f"Cohen's kappa: {cohen}")
print(f"Percent match: {per_match}")
```

```python
#Metamap Normalization Evaluation

tp = 0
fp = 0
fn = 0

filepath = '/Users/elizabethchang/Desktop/symbolic-project-data/interannotator'
file_list = os.listdir(os.path.join(filepath, 'cindy'))

for file in file_list:
    golden = pd.read_csv(os.path.join(filepath, 'cindy', file))
    golden_cui = golden['cui'].dropna().to_list()

    metamap = pd.read_csv(os.path.join(filepath,'metamap-csv', file))
    metamap_cui = metamap['cui'].dropna().to_list()

    golden_counter = Counter(golden_cui)
    metamap_counter = Counter(metamap_cui)

    for cui in golden_counter:
        if cui in metamap_counter:
            if golden_counter[cui] > metamap_counter[cui]:
                tp += metamap_counter[cui]
                fn += (golden_counter[cui] - metamap_counter[cui])
            elif golden_counter[cui] < metamap_counter[cui]:
                tp += golden_counter[cui]
                fp += (metamap_counter[cui] - golden_counter[cui])
            elif golden_counter[cui] == metamap_counter[cui]:
                tp += golden_counter[cui]
        else:
            fn += golden_counter[cui]
    for cui in metamap_counter:
        if cui not in golden_counter:
            fp += metamap_counter[cui]


print(f"Total TP: {tp}")
print(f"Total FP: {fp}")
print(f"Total FN: {fn}")

precision = tp / (tp + fp)
recall = tp / (tp + fn)
f = 2 * (precision * recall) / (precision + recall)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f}")
```
